```python
directory = 'C:/Users/18593/NDLAS A Primary/'
file = 'NLDAS_FORA0125_H.A20200308.0200.002.grb.SUB.nc4'

# use os.path.join to combine the parts
full_path = os.path.join(directory, file)
```

```python
tornado_raw = pd.read_csv("https://raw.githubusercontent.com/RouseJo/DTSC691/main/tornados.csv")
tornado_raw
```

```python
dt = pd.to_datetime(tornado_raw['datetime_utc'])

nomin = dt.apply(lambda t: t.replace(minute=0)).astype(str)
nomin
```

```python
tornado_raw['hr'] = nomin
tornado_raw
```

This is a selection of data between January 1st 2010 and December 31st 2020. New rating scales were used beginning in2007.

```python
tornado_decade = tornado_raw[(tornado_raw['yr'] > 2009) & (tornado_raw['yr'] < 2021)]
tornado_decade
```

```python
tornado_decade['hrcode'] = tornado_decade.hr.str.replace('-','')
tornado_decade
```

```python
tornado_decade['code'] =  tornado_decade['hrcode'].astype(str).str[:8] + '.' + tornado_decade['hrcode'].astype(str).str[9:11]

tornado_decade
```

```python
head = tornado_decade.head()
head
```

```python
dir_list = os.listdir(directory)

my_list = []
for dtg in tornado_decade['code']:
    for file_name in dir_list:
        if dtg in file_name:
            my_list.append(file_name)
            #print(file_name)
            break
    else:
        #print('na')
        my_list.append('None')

my_list
```

```python
tornado_decade['fn'] = my_list
tornado_decade
```

```python
filtered = tornado_decade.loc[tornado_decade['fn'] != 'None']
filtered
```

```python
df = pd.DataFrame(columns=["TMP", "SPFH", "PRES", "UGRD", "VGRD", "DLWRF", "CONVfrac", "CAPE", "PEVAP", "APCP", "DSWRF"]) 

for i in filtered['fn']:
    if i != 'none':
        sel_file = i
        j = filtered['slat'].iloc[0]
        k = filtered['slon'].iloc[0]
        sel_path = os.path.join(directory, sel_file)
        read_it = xr.open_dataset(sel_path, engine='netcdf4')
        isolate_it = read_it.sel(lat=j, lon=k, method='nearest').to_dataframe()
        features = isolate_it[["TMP", "SPFH", "PRES", "UGRD", "VGRD", "DLWRF", "CONVfrac", "CAPE", "PEVAP", "APCP", "DSWRF"]]
        feat_drop = features.reset_index(drop=True)
        #df = df.append(merge)
        df = pd.concat([feat_drop, df], ignore_index=True)
df
```

```python
merge = pd.concat([filtered.reset_index(drop=True), df.reset_index(drop=True)], axis=1) 
#merge = pd.concat([no_none, df], axis = 1)
merge
```
